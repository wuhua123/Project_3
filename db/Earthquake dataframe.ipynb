```python
# Take only the necessary columns
Earth_NOAA = earthquake_NOAA[['YEAR', 'COUNTRY', 'STATE','location3','LATITUDE', 'LONGITUDE', 'EQ_PRIMARY', 'FOCAL_DEPTH']]
print(Earth_NOAA.head())
```

```python
# Clear data and take only the recent 100 years data
Earth_NOAA_100 = Earth_NOAA.loc[Earth_NOAA["YEAR"] >= 1919, :]

Earth_NOAA_100 = Earth_NOAA_100.rename(columns={'YEAR':'year', 'COUNTRY':'country', 'STATE':'state','location3':'place', \
                                                'LATITUDE':'latitude', 'LONGITUDE':'longitude', 'EQ_PRIMARY':'mag', \
                                                'FOCAL_DEPTH':'depth'})

NOAA_100_years = Earth_NOAA_100[["year", "country", "state", "place", "latitude", "longitude", "mag", "depth"]]

NOAA_100_years[["country", "state"]] = NOAA_100_years[["country", "state"]].astype(str) 
for i in range(len(Earth_NOAA_100['state'])):
    if (Earth_NOAA_100.iloc[i,2]) != 'NaN':
        Earth_NOAA_100.iloc[i,2] = Earth_NOAA_100.iloc[i,1]
        Earth_NOAA_100.iloc[i,1] = "UNITED STATES"

USA = NOAA_100_years.loc[NOAA_100_years['country']=="UNITED STATES",:]
print(USA)
print('--------------------')

# get ride of 'state' and 'place' columns
NOAA_100 = NOAA_100_years[["year", "country", "latitude", "longitude", "mag", "depth"]]

print(NOAA_100.head())
print(NOAA_100.tail())
```
